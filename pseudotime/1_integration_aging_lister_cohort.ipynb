# Aging cohort + Lister cohort

## Parameter tweaking

```python
%%sh
dataPATH=/sc/arion/projects/CommonMind/aging/hui

for down_sample in 1000 2000 no
do
    for batch_key in no Batch
    do
        for n_hvg in 6000 4000 2000
        do
            for sub_pc in half 50
            do
                for n_neighbor in 25 50 100
                do
                    bash ~/submitMinerva.sh -Q premium \
                        -N 5 \
                        -W 12:00 \
                        -M 150000 \
                        -E snRNAseq_py3.9 \
                        -C "python ${dataPATH}/scripts/whole_adata_analysis.py \
-i ${dataPATH}/files/freeze3/240201_PsychAD_freeze3_AGING_284_pc_euch.h5ad@${dataPATH}/files/lister_processed.h5ad \
-b Aging@Lister \
-o ${dataPATH}/files/freeze3/aging_lister_combined_ds${down_sample}_rm${batch_key}_scanpy${n_hvg}_${sub_pc}_${n_neighbor} \
-d ${down_sample} \
-v scanpy@${n_hvg} \
-p ${sub_pc} \
-n ${n_neighbor} \
-k ${batch_key}" \
                        -J "${dataPATH}/logs/freeze3/aging_lister_combined_ds${down_sample}_rm${batch_key}_scanpy${n_hvg}_${sub_pc}_${n_neighbor}"
    
                    bsub < ${dataPATH}/logs/freeze3/aging_lister_combined_ds${down_sample}_rm${batch_key}_scanpy${n_hvg}_${sub_pc}_${n_neighbor}.sh
                done # for n_neighbor end
            done # for sub_pc end
        done # for n_hvg end
    done # for batch_key end
done # for down_sample end
```
